```python
import seaborn as sns
```

```python
reviews[reviews['price'] < 200].sample(1000).plot.hexbin(x='price', y='points', gridsize=15)
sns.kdeplot(reviews[reviews['price'] < 200].loc[:, ['price', 'points']].dropna().sample(1000))
```

```python
#sns.kdeplot(reviews[reviews['price']<200]['price'])
sns.kdeplot(reviews[reviews['price']<200].loc[:, ['price', 'points']].sample(1000))
```

```python
sns.jointplot(x='price', y='points', data=reviews[reviews['price'] < 100], kind='hex')
```

```python
df = reviews[reviews.variety.isin(reviews.variety.value_counts().head(5).index)]
```

```python
sns.boxplot(x='variety', y='points', data=df)
```
